```python
product_dataset["tags"]=product_dataset["about"]+product_dataset["category_name"]
```

```python
product_dataset=product_dataset.drop(columns=["about","category_name"])
```

```python
product_dataset
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=3000,stop_words='english')
cv
```

```python
vector = cv.fit_transform(product_dataset["tags"].values.astype('U')).toarray()
vector.shape
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
similarity
```

```python
product_dataset[product_dataset['name']=="Aquamarine Cargo Pants"].index[0]
```

```python
product_dataset.info()
```

```python
similarity[4]
```

```python
Distance = sorted(list(enumerate(similarity[4])), key=lambda vector: vector[1], reverse=True)
for i in Distance[0:5]:
    print(product_dataset.iloc[i[0]]['name'])
```

```python
def recommand(product):
    index = product_dataset[product_dataset['name']==product].index[0]
    Distance = sorted(list(enumerate(similarity[index])), key=lambda vector: vector[1], reverse=True)
    for i in Distance[1:6]:
        print(product_dataset.iloc[i[0]]['name'])
```

```python
recommand("Denesik Group Sneakers")
```

```python
import pickle 
pickle.dump(product_dataset,open('product_dataset.pkl','wb'))
pickle.dump(similarity,open('similarity.pkl','wb'))
```
